```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///../Databoston.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
# this doesn't work unless table in DB has primary keys.  
Base.classes.keys()
```

```python
crime = Base.classes.crime
codes = Base.classes.codes
```

# # # Working with the data, I decided to work on the above and go with Boston Crime data.


```python
incidents_per_month_by_year = """SELECT  
                                cr.YEAR as year,
                                COUNT(cr.MONTH) as crime_total_per_year
                                FROM crime as cr
                                GROUP BY cr.YEAR"""

df1 = pd.read_sql_query(incidents_per_month_by_year, engine)
df1
```

```python
df1.plot(kind="bar", x="year", y="crime_total_per_year", title="Total Crime Per Year Reported in Boston")
plt.savefig("../images/year.jpg")
```

```python
total_crimes_by_hour = """ SELECT  
                        COUNT (cr.HOUR) as total_crime_reported_by_hour, HOUR as hour, YEAR as year
                        FROM crime as cr
                        GROUP bY HOUR, year
"""
df2 = pd.read_sql_query(total_crimes_by_hour, engine)
df2.head(15)
```

```python
plt.scatter(df2.total_crime_reported_by_hour, df2.hour, c=df2.year)
plt.savefig("../images/hour.jpg")
```

```python
total_offenses = """SELECT OFFENSE_CODE_GOUP as offense, count(OFFENSE_CODE_GOUP) as sum_of_offense, YEAR
                    FROM Crime
                    WHERE(YEAR = '2018')
                    GROUP by OFFENSE_CODE_GOUP
                    ORDER BY sum_of_offense DESC"""
df3 = pd.read_sql_query(total_offenses, engine)
df3.head()
```

```python
df3.plot(kind="bar", x="offense", y="sum_of_offense", figsize = (8,4), title="Total Crime by Offense Type for 2018")
plt.savefig("../images/offense_type_2018.jpg")
```

```python
total_crime_by_district = """SELECT cr.DISTRICT as district, 
                                count(cr.ID) as total_offenses
                                FROM crime as cr
                                GROUP BY 1
                                ORDER BY 2 DESC
"""
df4 = pd.read_sql_query(total_crime_by_district, engine)
df5 = df4.fillna(0)
df5
```

```python
df5.plot(kind="bar", x="district", y="total_offenses", title="Total Offenses per District")
plt.savefig("../images/district.jpg")
```
